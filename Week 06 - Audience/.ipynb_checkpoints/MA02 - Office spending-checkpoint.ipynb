## Member office data
Get the locations of the previous "MemberData" for each Congress from the Internet Archive's Wayback Machine.

```python
responses_d = {}
responses_d['2015'] = requests.get('http://archive.org/wayback/available?url=https://clerk.house.gov/xml/lists/MemberData.xml&timestamp=20150201').json()
responses_d['2017'] = requests.get('http://archive.org/wayback/available?url=https://clerk.house.gov/xml/lists/MemberData.xml&timestamp=20170201').json()
responses_d['2019'] = requests.get('http://archive.org/wayback/available?url=https://clerk.house.gov/xml/lists/MemberData.xml&timestamp=20190201').json()
responses_d['2021'] = requests.get('http://archive.org/wayback/available?url=https://clerk.house.gov/xml/lists/MemberData.xml&timestamp=20210201').json()
```

Get the content of these MemberData fields.

```python
# Container
content_dfs = {}

# String for retrieving content
content_string = 'https://web.archive.org/web/{0}if_/https://clerk.house.gov/xml/lists/MemberData.xml'

# Loop through each year
for _year, _response in responses_d.items():
    
    # Pull timestamp
    _timestamp = _response['archived_snapshots']['closest']['timestamp']
    
    # Retrieve content
    _content = requests.get(content_string.format(_timestamp)).text
    
    # Parse XML into a DataFrame
    _df = pd.read_xml(_content,xpath='./members/*/member-info')
    
    # Extract Congress number
    _df['congress-num'] =  BeautifulSoup(_content).find_all('congress-num')[0].text
    
    content_dfs[_year] = _df
```

```python
# Clean up index
member_data_df = pd.concat(content_dfs,names=['year']).reset_index(level=0).reset_index(drop=True)

# Concat the office-building and office-room
member_data_df['office-full'] = member_data_df['office-building'] + ' ' + member_data_df['office-room'].astype('str')

member_data_df.head()
```

If members are not incumbents or haven't served in previous congresses, they have a 0.

```python
member_data_df['prior-congress'].value_counts()
```

```python
member_data_df['incumbent'] = member_data_df['prior-congress'].astype(int) == member_data_df['congress-num'].astype(int) - 1
```

Save to disk.

```python
member_data_df.to_csv('member_data_2015_2021.csv',index=False)
```

Make a map of bioguideIDs to official names.

```python
bioguide_name_map = member_data_df.set_index('bioguideID')['official-name'].to_dict()
bioguide_caucus_map = member_data_df.set_index('bioguideID')['caucus'].to_dict()
```

## Offices by year
Pivot data to get offices for each member by year.

```python
# Pivot to get officials' offices
member_offices_annual = pd.pivot_table(
    data = member_data_df,
    index = 'bioguideID',
    columns = 'year',
    values = 'office-full',
    aggfunc = 'sum'
).reset_index()

# Map official names from bioguideIDs
member_offices_annual['official-name'] = member_offices_annual['bioguideID'].map(bioguide_name_map)

# Inspect
member_offices_annual.head()
```

Melt data down.

```python
member_offices_annual_melt = pd.melt(
    member_offices_annual,
    id_vars = ['bioguideID','official-name'],
    var_name='year',
    value_name='office').sort_values(['bioguideID','year']).reset_index(drop=True)

# Drop missing values
member_offices_annual_melt = member_offices_annual_melt.dropna(subset=['office'])

# Inspect
print(member_offices_annual_melt.shape)
member_offices_annual_melt.head()
```

Are members incumbents?

```python
incumbents = pd.pivot_table(
    data = member_data_df,
    index = ['bioguideID','official-name'],
    columns = ['year'],
    values = 'incumbent',
    aggfunc = 'sum'
).stack().dropna().reset_index()

incumbents.rename(columns={0:'incumbent'},inplace=True)
```

Do members have the same office as the previous congress?

```python
gb_office = member_offices_annual_melt.groupby('bioguideID')['office']
member_offices_annual_melt['office-same'] = gb_office.shift(1) == gb_office.shift(0)
```

Join together.

```python
melted_office_df = pd.merge(
    left = member_offices_annual_melt,
    right = incumbents,
    left_on = ['bioguideID','year','official-name'],
    right_on = ['bioguideID','year','official-name'],
    how = 'left'
)

melted_office_df.head()
```

Aggregating and counting unique offices and congresses.

```python
# Define aggregation dictionary
agg_d = {'year':['nunique','min','max'],'office':'nunique'}

# Groupby-aggregation
member_years_offices = member_offices_annual_melt.groupby('bioguideID').agg(agg_d)

# Rename columns
member_years_offices.columns = ['congresses','min_year','max_year','unique_offices']

# Map in official names from bioguideIDs
member_years_offices['official-name'] = member_years_offices.index.map(bioguide_name_map)

# Inspect
member_years_offices.head()
```

```python
g = sb.catplot(
    data=member_years_offices,
    x='congresses',
    y='unique_offices',
    color='tab:blue',
    kind = 'bar'
)

ax = g.axes.flatten()[0]
ax.set_ylim((0,2.4))
ax.set_yticks(range(0,3))
ax.set_xlabel('Terms')
ax.set_ylabel('Unique offices')
ax.set_title('Office moves for U.S. House members')
plt.tight_layout()
plt.savefig('office_moves_terms.png',facecolor='w',bbox_inches='tight')
```

```python
g = sb.catplot(
    data = member_years_offices,
    x = 'min_year',
    y = 'unique_offices',
    kind = 'bar',
    order = year_order,
    color = 'tab:green'
)

ax = g.axes.flatten()[0]
ax.set_ylim((0,2.4))
ax.set_yticks(range(0,3))
ax.set_xlabel('First term')
ax.set_ylabel('Unique offices')
ax.set_title('Office moves for by term')
plt.tight_layout()
plt.savefig('office_moves_first_term.png',facecolor='w',bbox_inches='tight')
```

### Edgelist

```python
transitions_15_17 = member_offices_annual.copy()[['bioguideID','2015','2017']]
transitions_15_17.columns = ['bioguideID','from','to']
transitions_15_17['year'] = '2015->2017'
transitions_17_19 = member_offices_annual.copy()[['bioguideID','2017','2019']]
transitions_17_19.columns = ['bioguideID','from','to']
transitions_15_17['year'] = '2017->2019'
transitions_19_21 = member_offices_annual.copy()[['bioguideID','2019','2021']]
transitions_19_21.columns = ['bioguideID','from','to']
transitions_19_21['year'] = '2019->2021'

transitions_el = pd.concat([transitions_15_17,transitions_17_19,transitions_19_21]).dropna(subset=['from','to'],how='any')
transitions_el.head()
```

```python
transitions_wel = transitions_el.groupby(['from','to']).agg({'year':'nunique'}).reset_index()

# Filter out 0-weights and self-loops
c0 = transitions_wel['year'] > 0
c1 = transitions_wel['from'] != transitions_wel['to']
transitions_wel = transitions_wel[c0 & c1]
transitions_wel.columns = ['source','target','weight']
transitions_wel.head()
```

```python
g = nx.from_pandas_edgelist(
    df = transitions_wel,
    edge_attr = True,
    create_using=nx.DiGraph
)
```

```python
lcc_nodes = max(nx.connected_components(g.to_undirected()), key=len)
lcc_g = nx.subgraph(g,lcc_nodes)
pos = nx.layout.spring_layout(g)
```

```python
len(lcc_g)/len(g)
```

```python
f,ax = plt.subplots(figsize=(6,6))

nx.draw_networkx(
    g,
    with_labels = False,
    node_size = 50,
    pos = pos,
    ax = ax
)

f.savefig('hairball.png',facecolor='w')
```

## Disbursement data

Use the ProPublica "[House Office Expenditure Data](https://projects.propublica.org/represent/expenditures)" databases.

```python
files = sorted([f for f in os.listdir('./disbursement_data/') if '.csv' in f])
```

```python
disbursement_dfs = {}

for f in files:
    _quarter = f[:6]
    _df = pd.read_csv(
        './disbursement_data/' + f,
        encoding='latin1',
        low_memory=False,
        thousands=','
    )    
    
    _df.columns = [col.strip() for col in _df.columns]
        
    # Rename column
    if 'SORT SUBTOTAL DESCRIPTION' in _df.columns:
        _df = _df.rename(columns={'SORT SUBTOTAL DESCRIPTION':'CATEGORY'})
        
    disbursement_dfs[_quarter] = _df
```

```python
# Concatenate CSV files
all_db_df = pd.concat(disbursement_dfs,names=['YEAR-QUARTER']).reset_index(level=0)

# Extract and over-write inconsistent QUARTER and YEAR values
all_db_df['QUARTER'] = all_db_df['YEAR-QUARTER'].str.slice(4)
all_db_df['YEAR'] = all_db_df['YEAR-QUARTER'].str.slice(0,4)

# Different category values
all_db_df.replace(
    {'CATEGORY':
     {'RENT, COMMUNICATION, UTILITIES':'RENT COMMUNICATION UTILITIES',
      'RENT  COMMUNICATION  UTILITIES':'RENT COMMUNICATION UTILITIES'}
    },
    inplace=True
)

# Write to disk
all_db_df.to_csv('all_disbursements_2015_2021.csv',index=False)

# Inspect
print(all_db_df.shape)
all_db_df.head()
```

## Filtering data
Filtering data down to bioguideIDs.

```python
c0 = all_db_df['BIOGUIDE_ID'].notnull()
c1 = ~all_db_df['SORT SEQUENCE'].astype(str).str.contains('TOTAL')
bioguide_df = all_db_df.copy()[c0 & c1]
bioguide_df['AMOUNT'].fillna(0,inplace=True)

# Extract year and quarter as separate values
bioguide_df['YEAR'] = bioguide_df['YEAR-QUARTER'].str.slice(0,4)
bioguide_df['QUARTER'] = bioguide_df['YEAR-QUARTER'].str.slice(4)

print(bioguide_df.shape)
bioguide_df.head()
```

Pivot to members on index, quarters on columns, and total spending.

```python
quarterly_spending_df = pd.pivot_table(
    data = bioguide_df,
    index = ['BIOGUIDE_ID','YEAR-QUARTER'],
    columns = 'CATEGORY',
    values = 'AMOUNT',
    aggfunc = 'sum',
)

quarterly_spending_df.head(20)
```

```python
# Reshape
quarterly_spending_melt = quarterly_spending_df.stack().reset_index()

# Rename columns
quarterly_spending_melt.columns = ['BIOGUIDE_ID','YEAR-QUARTER','CATEGORY','AMOUNT']

# Add names of politicians
quarterly_spending_melt['NAME'] = quarterly_spending_melt['BIOGUIDE_ID'].map(bioguide_name_map)

# Sort the DataFrame by bioguide name, year-quarter, and category
quarterly_spending_melt.sort_values(['BIOGUIDE_ID','YEAR-QUARTER','CATEGORY'],inplace=True)

# Extract year and quarter as separate values
quarterly_spending_melt['YEAR'] = quarterly_spending_melt['YEAR-QUARTER'].str.slice(0,4)
quarterly_spending_melt['QUARTER'] = quarterly_spending_melt['YEAR-QUARTER'].str.slice(4)

# Inspect
quarterly_spending_melt.head()
```

```python
# Filter out rare categories
_df = quarterly_spending_melt[~quarterly_spending_melt['CATEGORY'].isin(['BENEFITS TO FORMER PERSONNEL','PERSONNEL BENEFITS'])]

# Make a catplot
g = sb.catplot(
    data = _df,
    x = 'AMOUNT',
    y = 'CATEGORY',
    hue = 'QUARTER',
    kind = 'bar',
    hue_order = ['Q1','Q2','Q3','Q4'],
    aspect = 1.5,
    palette = 'tab10',
    orient = 'h'
)

# Flatten and extract axes object
ax = g.axes.flatten()[0]

# Update axes properties
ax.set_xscale('log')
ax.set_xlim((1e1,1e6))
ax.set_ylabel('')

plt.savefig('total_disbursements_by_category_and_quarter.png',facecolor='w')
```

## Congress spending

```python
bioguide_df.head()
```

```python
annual_spending_df = pd.pivot_table(
    data = bioguide_df,
    index = ['BIOGUIDE_ID','YEAR'],
    columns = 'CATEGORY',
    values = 'AMOUNT',
    aggfunc = 'sum',
)

annual_spending_df.reset_index(inplace=True)
annual_spending_df['congress'] = annual_spending_df['YEAR'].map(year_congress_mapping)

annual_spending_df['TOTAL'] = annual_spending_df[annual_spending_df.columns[2:-1]].sum(1)
```

```python
merged_df = pd.merge(
    left = annual_spending_df,
    right = melted_office_df,
    left_on = ['BIOGUIDE_ID','YEAR'],
    right_on = ['bioguideID','year'],
    how = 'left'
)

merged_df.fillna(method='ffill',inplace=True)

merged_df['caucus'] = merged_df['BIOGUIDE_ID'].map(bioguide_caucus_map)

merged_df.head()
```

### Democrats spend more on

```python
c_dem = merged_df['caucus'] == 'D'
c_rep = merged_df['caucus'] == 'R'
c_incumb = merged_df['incumbent'] == 1
c_newbie = merged_df['incumbent'] == 0
```

```python
category = 'TOTAL'
print(stats.ttest_ind(merged_df.loc[c_dem,category],merged_df.loc[c_rep,category]))
print(stats.ttest_ind(merged_df.loc[c_incumb,category],merged_df.loc[c_newbie,category]))

g = sb.catplot(
    data = merged_df,
    x = 'incumbent',
    y = category,
    hue = 'caucus',
    kind = 'bar',
    palette = sb.set_palette(['red','blue'])
)

ax = g.axes.flatten()[0]
ax.set_title('Disbursements for {0},\nby party and incumbency'.format(category.title()))
ax.set_ylabel('Disbursements (dollars)')
ax.set_xticklabels(['Non-incumbents','Incumbents'])
ax.set_xlabel('')
g._legend.set_title('Party')

plt.savefig('spending_{0}.png'.format('_'.join(category.lower().split(' '))),facecolor='w',bbox_inches='tight')
```
